## Step 1: Prepare the Data for Clustering

```python
from sklearn.preprocessing import StandardScaler
# features for clustering: experience_level, weight, session duration, gender, workout type, age

features = ["age", "gender_category", "weight", "session_duration(h)", "experience_level", 
                 'workout_type_Cardio', 'workout_type_HIIT', 'workout_type_Strength', 'workout_type_Yoga']

# 3. Combine all feature

# 4. Create your feature matrix
X = df_numerical[features]

# 5. Scale the features (important for K-means)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() # normalized using Z-score
X_scaled = scaler.fit_transform(X)
# Calculates the mean and standard deviation of each feature in your data


print("Data prepared for clustering with shape:", X_scaled.shape)
```

## Step 2: Find the Optimal Number of Clusters

The elbow method helps us determine how many clusters (K) to use by plotting the sum of squared distances (inertia) against different values of K and looking for the "elbow" point.

```python
# Import necessary libraries
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Calculate inertia (within-cluster sum of squares) for different K values
inertia = []
k_range = range(1, 11)  # Test K from 1 to 10

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, 'o-', markersize=8, markerfacecolor='blue')
plt.grid(True)
plt.xlabel('Number of Clusters (K)', fontsize=12)
plt.ylabel('Inertia', fontsize=12)
plt.title('Elbow Method for Optimal K', fontsize=14)
plt.xticks(k_range)

# Add annotations to make the plot more informative
for k, i in zip(k_range, inertia):
    plt.annotate(f'{i:.0f}', (k, i), textcoords="offset points", 
                 xytext=(0, 10), ha='center')

plt.tight_layout()
plt.show()
```

## Step 3: Apply K-means with 4 Clusters

```python
# Apply K-means with the optimal number of clusters (K=4)
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_scaled)

# Add cluster labels to your original dataframe
df_numerical['cluster'] = cluster_labels

# Check how many members are in each cluster
cluster_counts = df_numerical['cluster'].value_counts().sort_index()
print("Members in each cluster:")
print(cluster_counts)
```

## Step 4: Analyze the Clusters

Next, let's analyze what makes each cluster unique by looking at the cluster centers and summary statistics:

```python
# Get cluster centers in the scaled space
cluster_centers_scaled = kmeans.cluster_centers_

# Transform the centers back to the original scale
cluster_centers = scaler.inverse_transform(cluster_centers_scaled)

# Create a dataframe of cluster centers with feature names
centers_df = pd.DataFrame(cluster_centers, columns=features)
centers_df.index.name = 'Cluster'

# Calculate the average values for each feature in each cluster
cluster_profiles = df_numerical.groupby('cluster').agg({
    'experience_level': 'mean',
    'weight': 'mean',
    'session_duration(h)': 'mean',
    'age': 'mean',
    'gender_category': 'mean',  # Assuming 0=Female, 1=Male, this gives % male
})

# Add count of members in each cluster
cluster_sizes = df_numerical['cluster'].value_counts().sort_index()
cluster_profiles['Member Count'] = cluster_sizes.values
cluster_profiles['Member Percentage'] = (cluster_sizes.values / len(df) * 100).round(1)

# For workout types, calculate percentage who do each type
for workout in ['workout_type_Cardio', 'workout_type_HIIT', 
               'workout_type_Strength', 'workout_type_Yoga']:
    cluster_profiles[workout.replace('workout_type_', '') + ' %'] = \
        df_numerical.groupby('cluster')[workout].mean() * 100

# Round all numeric values for readability
cluster_profiles = cluster_profiles.round(2)

# Display the cleaned up summary
print("=== CLUSTER PROFILES ===")
print(cluster_profiles)
```

### Add Cluster Labels to Dataframe

```python
# Get the cluster labels from the kmeans object
cluster_labels = kmeans.labels_

# Add cluster labels to your original dataframe
df_numerical['cluster'] = cluster_labels

# Create descriptive names for each cluster
cluster_names = {
    0: "Cardio",
    1: "Strength Training", 
    2: "High Intensity Interval Training",
    3: "Yoga"
}

# Add cluster names to your dataframe
df_numerical['segment'] = df_numerical['cluster'].map(cluster_names)

# View the first few rows with cluster assignments and segment names
print(df_numerical[['experience_level', 'session_duration(h)', 'weight', 'age', 'cluster', 'segment']].head())
```
